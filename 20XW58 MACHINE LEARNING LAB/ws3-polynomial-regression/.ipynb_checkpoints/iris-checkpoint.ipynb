```python
# 20pw16
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

file = pd.read_csv(r"Z:/SEM 5/20XW58 MACHINE LEARNING LAB/ws3-polynomial-regression/datasets/iris.csv")
```

```python
SepalLength = file['SepalLengthCm']
SepalWidth = file['SepalWidthCm']
PetalLength = file['PetalLengthCm']
PetalWidth = file['PetalWidthCm']
```

```python
def polynomial_regression(x,y,deg):
    
    X = []
    for i in x:
        row=[]
        for j in range(0,deg+1):
            row.append(i**j)
        X.append(row)
    
#     xtxi calc
    X = np.array(X)
    X_tp = X.T
    XTX = X_tp.dot(X)
    XTXi = np.linalg.inv(XTX)
    
#     xty calc
    Y=np.array(y)
    Y=Y.T
    XTY=X_tp.dot(Y)
    
#     print(XTX,XTY)
    
    B = XTXi.dot(XTY)
    print(f"The B values : {B}")

    y_cap =[]
    
    for i in x:
        term=0
        for j in range(0,deg+1):
            term += (i**j)*B[j]
        y_cap.append(term)
    
#     print(y_cap)
    
    plt.scatter(x,y)
    plt.scatter(x,y_cap)
    plt.show()
    
#     measures

    SSE=0
    SSR=0
    y_bar=sum(y)/len(y)
    for i in range(0,len(y)):
        SSE += (y_cap[i]-y[i])**2
        SSR += (y_cap[i]-y_bar)**2
    
    print(f"SSE = {SSE} \n, SSR = {SSR}\n SST = {SSE+SSR}.")
    print(f"Co-efficient of Determination = {SSR/(SSE+SSR)}")
```

```python
SepalLength = np.array(SepalLength)
SepalWidth = np.array(SepalWidth)
PetalLength = np.array(PetalLength)
PetalWidth = np.array(PetalWidth)
```

```python
# Quadratic
polynomial_regression(SepalLength,SepalWidth,2)
polynomial_regression(SepalLength,PetalLength,2)
polynomial_regression(SepalLength,PetalWidth,2)
polynomial_regression(PetalLength,PetalWidth,2)
polynomial_regression(PetalLength,SepalWidth,2)
polynomial_regression(PetalLength,SepalLength,2)
```

```python
# Cubic
polynomial_regression(SepalLength,SepalWidth,3)
polynomial_regression(SepalLength,PetalLength,3)
polynomial_regression(SepalLength,PetalWidth,3)
polynomial_regression(PetalLength,PetalWidth,3)
polynomial_regression(PetalLength,SepalWidth,3)
polynomial_regression(PetalLength,SepalLength,3)
```
