```python
def linear_regression(X,y,z):
    x=[]
    for i in X:
        x.append([1,i])
        
#   xTx inverse calc
    x=np.array(x)
    x_tp=x.T
    xtx=x_tp.dot(x)
    
    xtxi=np.linalg.inv(xtx)
    
#     xTy calc
    xty=x_tp.dot(y)
    
#     slope , intercept calc
    B=xtxi.dot(xty)
    
    b0=B[0]
    b1=B[1]
    print(f"slope {b1} xintercept {b0}")
#     generating y values
    y_cap=[]
    for i in X:
        y_cap.append(b0+b1*i)
    
#     comparing the values
    scatterPlot(X,y)
    plt.plot(X,y_cap)
    plt.xlabel(z)
    plt.ylabel("price")
    plt.show()

#   x vs price
linear_regression(bedrooms,price,"bedrooms")
linear_regression(sqft_living,price,"sqft_living")
linear_regression(sqft_lot,price,"sqft_lot")
linear_regression(sqft_above,price,"sqft_above")
linear_regression(sqft_basement,price,"sqft_basement")
```
