```python
# -*- coding: utf-8 -*-
"""Created on Tue Jul 19 14:51:48 2022@author: 20pw16"""

import pandas as pd
import numpy as np
import seaborn as sns
file = pd.read_csv(r"Z:/SEM 5/20XW58 MACHINE LEARNING LAB/ws1/datasets/iris.csv")

# sepalLength box plot for classes
sns.catplot(x="Species",y="SepalLengthCm",data=file,kind="box",col="Species")
sns.catplot(x="Species",y="SepalWidthCm",data=file,kind="box",col="Species")
sns.catplot(x="Species",y="PetalLengthCm",data=file,kind="box",col="Species")
sns.catplot(x="Species",y="PetalWidthCm",data=file,kind="box",col="Species")
```

```python
def find_outliers(colname,species):
    if(species=="Iris-setosa"):
        arr=file[colname][:50]
    elif(species=="Iris-versicolor"):
        arr=file[colname][50:100]
    elif(species=="Iris-virginica"):
        arr=file[colname][100:150]
    else:
        print("Invalid Input")
        return
    q1 = np.percentile(arr,25,axis=0)
    q3 = np.percentile(arr,75,axis=0)
    IQR = q3-q1
    lower = q1-1.5*IQR
    upper = q3+1.5*IQR
    outliers=[]
#     print(arr)
    for i in arr:
        if i<lower or i>upper:
            outliers.append(i)

    return outliers

print(f"Outliers for Iris Setosa :\nSepalLengthCm:{find_outliers('SepalLengthCm','Iris-setosa')}\nSepalWidthCm:{find_outliers('SepalWidthCm','Iris-setosa')}\nPetalLengthCm:{find_outliers('PetalLengthCm','Iris-setosa')}\nPetalWidthCm:{find_outliers('PetalWidthCm','Iris-setosa')}")
print(f"Outliers for Iris versicolor :\nSepalLengthCm:{find_outliers('SepalLengthCm','Iris-versicolor')}\nSepalWidthCm:{find_outliers('SepalWidthCm','Iris-versicolor')}\nPetalLengthCm:{find_outliers('PetalLengthCm','Iris-versicolor')}\nPetalWidthCm:{find_outliers('PetalWidthCm','Iris-versicolor')}")
print(f"Outliers for Iris virginica :\nSepalLengthCm:{find_outliers('SepalLengthCm','Iris-virginica')}\nSepalWidthCm:{find_outliers('SepalWidthCm','Iris-virginica')}\nPetalLengthCm:{find_outliers('PetalLengthCm','Iris-virginica')}\nPetalWidthCm:{find_outliers('PetalWidthCm','Iris-virginica')}")

# find_outliers("SepalWidthCm","Iris-virginica")
```
