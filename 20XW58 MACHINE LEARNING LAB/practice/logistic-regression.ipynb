```python
import matplotlib.pyplot as plot
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("D:/PSG/PSG-Z/SEM 5/20XW58 MACHINE LEARNING LAB/practice/datasets/data.csv")
del df['id']
del df['Unnamed: 32']
```

```python
df.head()
```

```python
le = LabelEncoder()

df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head()
```

```python
features = df.columns[1:]
```

```python
def remove_multicollinearity(features):
    final_features = [x for x in features]
    
    CR = df[features].corr().values.tolist()
    
    for i in range(len(CR)):
        for j in range(i+1,len(CR)):
            if abs(CR[i][j])>0.7 and features[i] in final_features:
                final_features.remove(features[i])
    return final_features
```

```python
final_features = remove_multicollinearity(features)
```

```python
df[final_features]
```

```python
X = df[final_features]
Y = df['diagnosis']
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def logistic_regression(x,y):
    W = np.random.rand(x.shape[1]+1)
    eta = 0.5
    convergence = True
    y_cap = [None]*len(y)
    epochs = 0
    Wnew = np.random.rand(x.shape[1]+1)
    
    while(convergence):
        
        
        for i in range(len(x)):
            wTx = W.dot(np.insert(x[i],0,1))
 
            if(sigmoid(wTx)>=0.5):
                y_cap[i] = 1
            else:
                y_cap[i] = 0
                
            if(y_cap[i]!=y[i]):
                Wnew = W + eta*(y[i]-y_cap[i])*y_cap[i]*(1-y_cap[i])*np.insert(x[i],0,1)
            
            W = Wnew
    
        epochs += 1
        if(epochs==6000):    
            convergence = False
    
    return W
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X.to_numpy(),Y.to_numpy(),test_size=0.33)
```

```python
weights = logistic_regression(X_train,Y_train)
```

```python
def model_testing(W,X,Y):
    Y_cap = [None]*len(Y)
    X = list(X)
    for i in range(len(X)):
        wTx = W.dot(np.insert(X[i],0,1))
        
        if(sigmoid(wTx)>=0.5):
            Y_cap[i] = 1
        else:
            Y_cap[i] = 0
    
    return Y_cap
            
```

```python
Y_pred = model_testing(weights,X_test,Y_test)
```

```python
def metrics_check(y,y_cap):
    count = 0
    
    for i in range(len(y)):
        if(y[i]==y_cap[i]):
            count += 1
            
    return count/len(y)
```

```python
acc = metrics_check(Y_test,Y_pred)
```

```python
print(f"the accuracy of the model is : {acc}")
```
