```python
from sklearn import datasets
import numpy as np
import random
```

```python
def dist(x,y):
    dist = 0
    for i in range(len(x)):
        dist += (x[i]-y[i])**2
    return dist**0.5
```

```python
def KMeans(data,n_clusters=3):
    n = len(data[0])
    
    old_centroids = [data[random.randint(0,len(data)-1)] for i in range(n_clusters)]
    
    new_centroids = list(old_centroids)
    
    while True:
        labels = []
        count = [0]*n_clusters
        
        for i in range(len(data)):
            distance = []
            for j in range(n_clusters):
                distance.append(dist(data[i],old_centroids[j]))
            
            label = distance.index(min(distance))
            labels.append(label)
            count[label] += 1
            
        print(labels)
        
        for i in range(n_clusters):
            new_centroids[i] = [0]*n
            
        for i in range(len(data)):
            new_centroids[labels[i]] += data[i]/count[labels[i]]
            
        if(np.array_equal(new_centroids,old_centroids)):
            break
        else:
            old_centroids = new_centroids

    return {'labels':labels,'centroids':old_centroids,'count':count}
```

```python
iris = datasets.load_iris()
data = iris.data
```

```python
KMeans(data,3)
```
