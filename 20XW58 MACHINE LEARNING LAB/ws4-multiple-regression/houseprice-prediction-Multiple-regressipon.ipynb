```python
def scatterPlot(x,y):
    plt.scatter(x,y)
    plt.show()
```

```python
def multiple_regression(x_components,y):
    x=[]
    y=np.array(y)
    size=len(x_components[0])

    for i in range(0,size):
        x.append([x_components[x][i] for x in range(0,len(x_components))])
        x[i].insert(0,1)
    
    x=np.array(x)
    x_tp = x.T
#     print(x)
#     print(y)
    xTx = x_tp.dot(x)
#     print(xTx)
    
    xTy = x_tp.dot(y)
#     print(xTy)
    
    xTxi = np.linalg.inv(xTx)
    
    W = xTxi.dot(xTy)
    print(W)
    
    y_cap= []
    
    for i in range(0,size):
        yhat=W[0]
        for j in range(0,len(x_components)):
            yhat+=W[j+1]*x_components[j][i]
#             print(x_components[j][i])
        y_cap.append(yhat)
    
    print(y_cap)
#     scatterPlot(x_components[0],y)
#     scatterPlot(x_components[0],y_cap)
```
